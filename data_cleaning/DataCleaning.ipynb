```python
# Loop through 375 pages
for x in range(1,376):
    page_number = x

    # Endpoint for finding most popular movies
    most_popular_url = f"https://api.themoviedb.org/3/discover/movie?api_key={api_key}&page={page_number}&sort_by=popularity.desc"
    # Most popular movies
    tmdb_response = requests.get(most_popular_url).json()
    results = tmdb_response["results"]

    # Loop through the length of each page
    for y in range(20):
        movie_title = results[y]["title"]
        movie_id = results[y]["id"]
        
        # Endpoint & response
        movie_url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={api_key}"
        tmdb_response_2 = requests.get(movie_url).json()

        # Get keywords for each movie
        keywords_url = f"https://api.themoviedb.org/3/movie/{movie_id}/keywords?api_key={api_key}"
        keywords_response = requests.get(keywords_url).json()

        # pprint(tmdb_response)

        # Check budget & revenue
        film_budget = tmdb_response_2["budget"]
        film_revenue = tmdb_response_2["revenue"]
        film_words = keywords_response["keywords"]
        
        # If ID isn't already in list of IDs (filter out any duplicates)
        if movie_id not in movie_ids:
            
            # If budget & revenue exist in database
            if (film_budget != 0) & (film_revenue != 0):
                
                # If keywords exist in database
                if (film_words):
                    print(f"Page {page_number}: Appending {movie_title}")
                    
                    # Append all results
                    movie_ids.append(tmdb_response_2["id"])
                    titles.append(tmdb_response_2["title"])
                    budget.append(film_budget)
                    revenue.append(film_revenue)
                    vote_average.append(tmdb_response_2["vote_average"])
                    vote_count.append(tmdb_response_2["vote_count"])
                    release_date.append(tmdb_response_2["release_date"])
                    runtime.append(tmdb_response_2["runtime"])
                    popularity.append(tmdb_response_2["popularity"])
                    genres.append(tmdb_response_2["genres"])
                    spoken_languages.append(tmdb_response_2["spoken_languages"])
                    original_languages.append(tmdb_response_2["original_language"])
                    production_companies.append(tmdb_response_2["production_companies"])
                    production_countries.append(tmdb_response_2["production_countries"])
                    tagline.append(tmdb_response_2["tagline"])
                    overview.append(tmdb_response_2["overview"])
                    status.append(tmdb_response_2["status"])
                    keywords.append(film_words)
        
#pprint(tmdb_response)
    
```

### Function to Transfer Lists into Strings:

```python
# Function to turn list into string
def listToString(x):
    
    # Define a space variable
    space = " "
    
    # Return the joined list
    return(space.join(x))
```

### Function to Check for Blanks in Lists:

```python
def check_for_Blanks(x):
    
    # If x string is not blank
    if x and x.strip(): 
        return False    
    
    # If x string is blank
    return True 
```

## International Films:

### Find Number of Countries:

```python
# Number of Production Countries
num_production_countries = []

# Loop through to add number of countries per film
for x in range(len(production_co_origin_country)):
    # If country exists
    if production_co_origin_country[x]:
        num_production_countries.append(len(production_co_origin_country[x]))
    # If country is blank
    if not production_co_origin_country[x]:
        num_production_countries.append(0)
```

### Adjust Number of Countries (if blanks exist):

```python
# Loop through to get rid of blank strings in lists
for x in range(len(production_co_origin_country)):
    
    # Initialize a subtraction variable
    subtract = 0
    
    # Loop through each production company list
    for y in range(len(production_co_origin_country[x])):
        
        # Call function
        blanks = check_for_Blanks(production_co_origin_country[x][y])
        
        # If blanks exist, add to the subtraction variable
        if blanks == True:
            subtract+=1
            
    # If the number of countries is not 0
    if num_production_countries[x] != 0:
        num_production_countries[x] = num_production_countries[x] - subtract
```

### Find how many companies are based in US:

```python
# Create list
domestic = []

# Loop through to add number of companies based in US
for x in range(len(production_co_origin_country)):
    country = 0
    
    for y in range(len(production_co_origin_country[x])):
        if (production_co_origin_country[x]):
            # If not based in US
            if (production_co_origin_country[x][y]) != "US":
                country+=0
            # If based in US
            elif (production_co_origin_country[x][y]) == "US":
                country+=1
                
    # Append to domestic list 
    domestic.append(country)
```

### Calculate percentage of companies based in US for each movie:

```python
# Create list
percentage_us = []

# Loop through to append percentages
for x in range(len(domestic)):
    if (num_production_countries[x]) != 0:
        percentage_us.append(domestic[x] / num_production_countries[x])
    else:
        percentage_us.append(None)
```

## Cast:

```python
# Create lists
actors = []
genders = []

# Loop through to 
for x in range(len(cast_list)):
    actor = []
    gender = []
    
    # Loop through to append names & genders to lists
    for y in range(len(cast_list[x])):
        
        # Append to pre-lists
        actor.append(cast_list[x][y]["name"])
        gender.append(cast_list[x][y]["gender"])
    
    # Append to lists
    actors.append(listToString(actor))
    genders.append(gender)
```

```python
len(genders)
```

### Total Cast:

```python
num_actors = []
for x in range(len(cast_list)):
    num_actors.append(len(cast_list[x]))
# num_actors
```

### Total Female Cast:

```python
# Find number of females in cast
num_female_cast = []

# Number of Female Cast Members
for x in range(len(cast_list)):
    num_female = 0
    
    for y in range(len(cast_list[x])):
        if genders[x][y] == 1:
            num_female+=1
    num_female_cast.append(num_female)
        
# print(num_female_cast)
```

### Female Led Cast (Top 5):

```python
# Find number of females in cast
female_led = []
female_led_per = []

# Number of Female Cast Members
for x in range(len(cast_list)):
    num_female = 0
    total = 0
    
    # If cast is greater than or equal to 5 actors
    if (len(cast_list[x])) >= 5:
        for y in range(5):
            if genders[x][y] == 1:
                num_female+=1
            total+=1
            
    # If cast has less than 5 actors
    elif (len(cast_list[x]) < 5) & (len(cast_list[x]) > 0):
        for y in range(len(cast_list[x])):
            if genders[x][y] == 1:
                num_female+=1
            total+=1
    else:
        total+=1
            
    # If exception occurs
    try:
        # Append the percentage of female cast in top 5
        female_led_per.append(num_female / total)
        female_led.append(num_female)
        
    except ZeroDivisionError:
        print(f"Total Equals 0 at {x}")
        
        
# print(female_led)
```

### Percentage Female Cast:

```python
# Find percentage of female cast
percentage_female_cast = []

# Loop through to calculate
for x in range(len(num_actors)):
    
    # If exception occurs
    try:
        percentage_female_cast.append(num_female_cast[x] / num_actors[x])
    except ZeroDivisionError:
        percentage_female_cast.append(num_female_cast[x] / 1)

# print(percentage_female_cast)
```

## Crew:

```python
# Create lists
crew_members = []
crew_genders = []

directors = []
directors_gender = []
directors_female = []

# Iterate through to append to lists
for x in range(len(crew_list)):
    
    # Initialize pre-lists
    member = []
    gender = []
    director = []
    director_gender = []
    female_director = []
    
    
    # Iterate through to append to pre-lists
    for y in range(len(crew_list[x])):
        
        # Create job variable
        the_job = crew_list[x][y]["job"]
        person_gender = crew_list[x][y]["gender"]
        
        # Convert variable to string for list
        person_gender = str(person_gender)
        
        # Append values to pre-lists
        member.append(crew_list[x][y]["name"])
        gender.append(crew_list[x][y]["gender"])
        
        # If the crew's job is a director, append name & gender
        if the_job == "Director":
            director.append(crew_list[x][y]["name"])
            director_gender.append(person_gender)
            
            # Create a female director column
            if person_gender == "1":
                female_director.append(1)
            else:
                female_director.append(0)
                
    # Append pre-lists to actual lists
    crew_members.append(listToString(member))
    crew_genders.append(gender)

    # Append director information to lists
    directors.append(listToString(director))
    directors_gender.append(listToString(director_gender))
    directors_female.append(female_director)
```

### Percentage Female Directed: (if more than 1 Director)

```python
# Create list
percentage_female_directed = []

# Loop through to find multiple directors
for x in range(len(directors_female)):
    total_directors = len(directors_female[x])
    female_directed = 0
    
    for y in range(total_directors):
        
#         # If multiple directors
#         if (total_directors) > 1:

        if directors_female[x][y] == 1:
            female_directed+=1
        elif directors_female[x][y] == 2:
            female_directed+=0
            
#         # If just one director
#         else:
#             if directors_female[x][y] == 1:
#                 female_directed+=1
#             elif directors_female[x][y] == 2:
#                 female_directed+=0

    # Calculate percentage of female directors
    percentage_female_directed.append(female_directed / total_directors)
                
```
